```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("listings.csv")
df.head()
```

```python
df.shape
df.info()
df.describe(include='all').transpose()
```

```python

print(df.isnull().sum())
df = df.drop_duplicates()

df['price'] = df['price'].fillna(df['price'].median())

df.info()
```

```python
sns.countplot(x='room_type', data=df, palette='pastel')
plt.title("Distribution of Room Types")
plt.xlabel("Room Type")
plt.ylabel("Number of Listings")
plt.show()
#This plot shows the frequency of each room type, which helps understand what Airbnb guests in Singapore prefer.
```

```python
sns.histplot(df['price'], bins=30, kde=True, color='teal')
plt.title("Distribution of Listing Prices")
plt.xlabel("Price (SGD)")
plt.ylabel("Number of Listings")
plt.xlim(0, 1000)
plt.show()
```

```python
top_neighbourhoods = df['neighbourhood'].value_counts().head(10)
top_neighbourhoods.plot(kind='bar', color='orange')
plt.title("Top 10 Neighbourhoods by Number of Listings")
plt.xlabel("Neighbourhood")
plt.ylabel("Listing Count")
plt.show()
#Shows which neighbourhoods are most popular for Airbnb listings.
```

```python
sns.scatterplot(x='minimum_nights', y='price', data=df, alpha=0.6)
plt.title("Minimum Nights vs Price")
plt.xlabel("Minimum Nights")
plt.ylabel("Price (SGD)")
plt.show()
#Scatter shows relationship between minimum stay and price.
```

```python
sns.boxplot(x='room_type', y='price', data=df, palette='coolwarm')
plt.title("Price Distribution by Room Type")
plt.xlabel("Room Type")
plt.ylabel("Price (SGD)")
plt.ylim(0, 1000)
plt.show()
```

```python
sns.histplot(df['number_of_reviews'], bins=30, kde=False, color='purple')
plt.title("Distribution of Number of Reviews")
plt.xlabel("Number of Reviews")
plt.ylabel("Frequency")
plt.show()
```
