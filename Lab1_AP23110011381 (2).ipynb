## <a name='reshape'></a>5Ô∏è‚É£ Reshaping, Transpose & Copies vs‚ÄØViews

```python
b = np.arange(8)
B = b.reshape(2,4)
B[0,0] = 99
print('b is modified:', b)
C = b.reshape(2,4).copy()
C[0,0] = -1
print('b unchanged with copy:', b)
```

```python
# üñäÔ∏è TODO: Flatten a 3‚ÄëD array into 1‚ÄëD using both `ravel` and `flatten`; observe copy vs view.
a = np.arange(24).reshape(2, 3, 4)
print(a)
r = a.ravel()
print(r)
f = a.flatten()
print(f)
r[0] = 999
print(a)
f[1] = 555
print(a)
```

## <a name='project'></a>1Ô∏è‚É£6Ô∏è‚É£ Mini‚ÄëProject: Fitness Data Analysis

Load `fitness.txt` (tab‚Äëseparated) then follow prompts.

```python
fitness = np.genfromtxt('fitness.txt', delimiter='\t', dtype=None, encoding=None, names=True)
print('columns:', fitness.dtype.names, 'rows:', len(fitness))
```

```python
# üñäÔ∏è TODO: Monthly step count, sleep vs mood correlation, weekly summary, etc.
dates = fitness['date']
steps = fitness['step_count']
sleep = fitness['hours_of_sleep']
mood = fitness['mood']

mons = np.array([d[3:] for d in dates])
uniq_mons = np.unique(mons)

mon_steps = {}
for m in uniq_mons:
    mask = mons == m
    mon_steps[m] = np.sum(steps[mask])

print("\n Monthly Step Count:")
for m, total in mon_steps.items():
    print(f"  {m}: {total} steps")

corr = np.corrcoef(sleep, mood)[0, 1]
print(f"\n Correlation between sleep and mood: {corr:.3f}")

length = len(fitness)
full_wks = length // 7

wk_steps = np.array_split(steps[:full_wks * 7], full_wks)
wk_sleep = np.array_split(sleep[:full_wks * 7], full_wks)
wk_mood = np.array_split(mood[:full_wks * 7], full_wks)

print("\nWeekly Summary:")
for i in range(full_wks):
    print(f"  Week {i+1}: steps={np.mean(wk_steps[i]):.0f}, "
          f"sleep={np.mean(wk_sleep[i]):.1f}, mood={np.mean(wk_mood[i]):.0f}")
```
